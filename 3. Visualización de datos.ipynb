```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #unused
```

<h2>Gr√°fico de barras<h2/>

```python
films = ['Film A', 'Film B', 'Film C', 'Film D', 'Film E']
sales = [1500000, 2200000, 1800000, 2800000, 1200000]  # in dollars

# Creaci√≥n del plot
plt.figure(figsize=(10, 6))  # opcional
plt.bar(films, sales, color='skyblue')

# se agrega labels a los ejes y t√≠tulo al plot
plt.xlabel('Films')
plt.ylabel('Sales ($)')
plt.title('Film Sales')

# Mostrar Plot
plt.show()
```

<h2>Gr√°fico de l√≠neas<h2/>

```python
# a√±os
years = [2018, 2019, 2020, 2021, 2022]

# precios
film_a_sales = [1200000, 1400000, 1800000, 2100000, 2500000]
film_b_sales = [1800000, 2100000, 2400000, 2600000, 3000000]

# Crear gr√°fico
plt.figure(figsize=(10, 6))  # Opcional: ajustar el tama√±o de la figura
plt.plot(years, film_a_sales, marker='o', label='Film A')
plt.plot(years, film_b_sales, marker='o', label='Film B')

# titulo y labels
plt.xlabel('Year')
plt.ylabel('Sales ($)')
plt.title('Film Sales Over Time')
plt.legend()  # Display legend to differentiate between films

# Show the plot
plt.grid(True)  # opcional: se agrega cuadr√≠cula
plt.tight_layout()  # mejorar el espaciado
plt.show()
```

<h2>Gr√°fico de dispersi√≥n<h2/>

```python
# data
films = ['Film A', 'Film B', 'Film C', 'Film D', 'Film E']
budgets = [80000000, 95000000, 120000000, 60000000, 75000000]
earnings = [180000000, 220000000, 260000000, 140000000, 190000000]

# crear plot
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size
plt.scatter(budgets, earnings, color='blue', marker='o')

# labels y t√≠tulo
plt.xlabel('Budget ($)')
plt.ylabel('Box Office Earnings ($)')
plt.title('Film Budget vs Box Office Earnings')

# Agregar nombres de pel√≠culas como etiquetas
for i, film in enumerate(films):
    plt.annotate(film, (budgets[i], earnings[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Mostrar
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h2>Gr√°fico de area circular<h2/>

```python
from math import pi

plt.figure(figsize=(7,9))

# Datos
df = pd.DataFrame({
'grupos': ['ordenes', 'ventas'],
'Ene': [105, 111],
'Feb': [50, 47],
'Mar': [123, 165],
'Abr': [211, 203], 
'May': [184, 110], 
'Jun': [180, 105],
'Jul': [183, 103], 
'Ago': [175, 106], 
'Sep': [101, 130], 
'Oct': [106, 120], 
'Nov': [146, 140], 
'Dic': [190, 206],
})

# N√∫mero de variables
meses=list(df)[1:] # ['Ene', 'Feb', 'Mar', ..., 'Dic']
N = len(meses) # 12

# Cu√°l va a ser el √°ngulo entre cada eje en el gr√°fico ? (dividimos el gr√°fico / el n√∫mero de variables)
angulos = [n / float(N) * 2 * pi for n in range(N)] # [0.0, 0.52, 1.04, ..., 0.0 ] # radianes
angulos += angulos[:1] # [0.52, 1.04, ..., 0.0 ]

# Inicializar el diagrama radial o de ara√±a
ax = plt.subplot(111, polar=True)

# Si quieres que el primer eje est√© en la parte superior:
ax.set_theta_offset(pi / 2) # 90¬∞
ax.set_theta_direction(-1) # Hacia la izquierda

# Dibuja un eje por categor√≠a + agrega etiquetas a√∫n
plt.xticks(angulos[:-1], meses)

# Dibuja los ejes "y"
ax.set_rlabel_position(0)
plt.yticks([50,100,150,200], ["50","100","150", "200"], color="grey", size=12)
plt.ylim(0,250)

# Gr√°fico individual = cada l√≠nea de los datos

# √ìrdenes
values=df.loc[0].drop('grupos').values.flatten().tolist() # [105, 50, 123, ..., 190]
values += values[:1] # [105, 50, 123, ..., 190, 105]
ax.plot(angulos, values, linewidth=3, linestyle='solid', label="√ìrdenes")
ax.fill(angulos, values, 'b', alpha=0.1)

# Ventas
values=df.loc[1].drop('grupos').values.flatten().tolist()
values += values[:1]
ax.plot(angulos, values, linewidth=3, linestyle='solid', label="Ventas")
ax.fill(angulos, values, 'r', alpha=0.1)

# Agregar leyenda
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```

<h2>Gr√°fico de burbujas<h2/>

```python
# data
x_values = np.random.rand(20) * 100
y_values = np.random.rand(20) * 100

bubble_sizes = np.random.rand(20) * 800  # tama√±o de las burbuejas
labels = ['Label {}'.format(i+1) for i in range(20)]


plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, s=bubble_sizes, c='blue', alpha=0.5)

#Agregar etiquetas a las burbujas
for i, label in enumerate(labels):
    plt.annotate(label, (x_values[i], y_values[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Bubble Chart Example')

plt.grid(True)
plt.tight_layout()
plt.show()
```

<h2>Boxplot<h2/>

```python
# Sample data for the box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a box plot
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
plt.boxplot(data, vert=True, patch_artist=True)

# Adding labels and title
plt.xticks([1, 2, 3], ['Category 1', 'Category 2', 'Category 3'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.grid(True)  # Optional: Add gridlines for better readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
