```python
import matplotlib as plt

# cuentas (pandas.core.series.Series)
discriminacion = df['booking_status'].value_counts()
print(discriminacion)

# conteo de totales
percent = df.shape[0]
print('\ntotal de bookings: ',percent)

# porcentaje (pandas.core.series.Series)
discriminacion / percent

discriminacion.plot(kind='bar')
```
